# Question 3

Our project aims to analyse trends in data careers from 2017 to 2022, primarily using six datasets obtained from Kaggle. These datasets comprise survey answers related to data careers, serving as our primary information source. However, we've identified potential issues with the datasets. 

First, different questions were asked across the years, resulting in variations in the column structures across the datasets.  To deal with this, we will clean the dataset of each year, dropping appropriate columns and re-structuring the datasets as needed prior to appending them. 

Second, some columns measures are captured differently across the years. For example, in earlier years, answers to how often certain tools, methods and platforms are used are captured on a 4 point likert scale -  Rarely, Sometimes, Often and Most of the time. In later years, the question is asking which tools, methods and platforms are regularly used, captured as binary measures - Yes or No. To address this,  we have created a new variable for the answers in the earlier years, and re-coded the Often and Most of the time answers as Yes,  and the Rarely, Sometimes answers as No. This is because we can reasonably deduce the first two answers as equating to regularly. 

Third, the datasets multiple choice variables are not coded as 0 and 1, but rather with the name of a tool or with a null value, which will complicate our analyses when seeking to understand product popularity. To address this we will first check if any multiple choice question has gone completely unanswered, drop the rows of those that have been, and recode the answers as 0 or 1 accordingly.

The fourth problem with our survey is the subjectivity involved in the survey answers, particularly in self-reported frequency of tools usage, categorised as "Rarely", "Sometimes", "Often", and "Most of the time". For instance, an individual who self-reports using TPUs "regularly" recorded usage between 15 to 25 times. Contrarily, another individual self-reporting as using TPUs "rarely" showed usage exceeding 25 times. These cases demonstrate the lack of quantitative standards for frequencies in our dataset. However, this doesn't significantly impact our core objective of providing valuable insights to prospective data professionals. Our focus is on capturing overarching trends, not the exact frequencies of tools usage. We believe that despite the subjectivity, the responses still offer a reliable gauge of tool usage trends in the data community.

The fifth issue we need to consider is the bias in the dataset. Our dataset originates from surveys conducted on Kaggle, introducing a potential source of bias. As people who took this survey are all Kaggle users, our dataset cannot comprehensively represent the broader population of data scientists/analysts/engineers. Particularly, it might overlook data professionals who do not have access to Kaggle or choose not to use it. This aspect could limit the universality of our findings, making them more reflective of trends within the Kaggle community rather than the entire field of data science. While we acknowledge that our dataset may not capture the experiences of every data professional, especially those without access to Kaggle, the large and diverse user base of Kaggle still provides a rich, varied and representative subset of the data science community. The trends, tool usage patterns, and career paths observed within this subset can still offer meaningful and actionable insights for both aspiring and current data science professionals.

Additionally, to enhance our insights, we'll use three APIs. The first, Google Trends will help gauge interest in data careers and tools over time and across countries. The second, Google Translate API will enable us to conduct Google Trend searches in different languages, ensuring data accuracy. Additionally, we'll use a third API to adjust for salary inflation over time and across various countries.
